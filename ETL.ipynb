### Comenzamos importando la librerias necesarias para el trabajo.

```python
#Importo las librerias para leer archivos

import pandas as pd
import numpy as np
```

```python
# Importamos los archivos csv.

amazon=pd.read_csv('./MLOpsReviews/amazon_prime_titles.csv')
disney=pd.read_csv('./MLOpsReviews/disney_plus_titles.csv')
hulu=pd.read_csv('./MLOpsReviews/hulu_titles.csv')
netflix=pd.read_csv('./MLOpsReviews/netflix_titles.csv')
```

```python
# Creación del campo id en las variables de nuestros archivos.
amazon['id'] = amazon['show_id'].apply(lambda x: 'a' + str(x))
disney['id'] = disney['show_id'].apply(lambda x: 'd' + str(x))
hulu['id'] = hulu['show_id'].apply(lambda x: 'h' + str(x))
netflix['id'] = netflix['show_id'].apply(lambda x: 'n' + str(x))
```

```python
# Genero columna con el nombre de cada plataforma
amazon['platform'] = 'amazon'
disney['platform'] = 'disney'
hulu['platform'] = 'hulu'
netflix['platform'] = 'netflix'
```

```python
#2- Reemplazar valores nulos en el campo rating por "G"
amazon['rating'].fillna('G', inplace=True)
disney['rating'].fillna('G', inplace=True)
hulu['rating'].fillna('G', inplace=True)
netflix['rating'].fillna('G', inplace=True)
```

```python
# 3-Convertir las fechas en formato AAAA-mm-dd
amazon['date_added'] = pd.to_datetime(amazon['date_added'])
disney['date_added'] = pd.to_datetime(disney['date_added'])
hulu['date_added'] = pd.to_datetime(hulu['date_added'])
netflix['date_added'] = pd.to_datetime(netflix['date_added'])
```

```python
# Convertimos los datos en minusculas.
amazon = amazon.applymap(lambda s: str(s).lower())
disney = disney.applymap(lambda s: str(s).lower())
hulu = hulu.applymap(lambda s: str(s).lower())
netflix= netflix.applymap(lambda s: str(s).lower())
```

```python
# Unificar el valor season 
amazon['duration'].replace('seasons','season', inplace=True, regex=True)
disney['duration'].replace('seasons','season', inplace=True, regex=True)
hulu['duration'].replace('seasons','season', inplace=True, regex=True)
netflix['duration'].replace('seasons','season', inplace=True, regex=True)
```

```python
# Separar la columna duration en 2 nuevas columnas duration_int y duration_type 
amazon[['duration_int','duration_type']]=amazon.duration.str.split(' ',expand=True)
disney[['duration_int','duration_type']]=disney.duration.str.split(' ',expand=True)
hulu[['duration_int','duration_type']]=hulu.duration.str.split(' ',expand=True)
netflix[['duration_int','duration_type']]=netflix.duration.str.split(' ',expand=True)
```

```python
#Cambiar el tipo de dato para la columna duration_int
amazon['duration_int'] = amazon['duration_int'].astype(int)
disney['duration_int'] = disney['duration_int'].astype(int)
```

```python
#Para cambiar el tipo de dato para la columna duration_int en hulu es necesario reemplazar los valores 'nan'
hulu['duration_int'].replace("nan","0", inplace=True, regex=True)
hulu['duration_int'] = hulu['duration_int'].astype(int)
```

```python
# Convertir la columna "duration" a tipo numérico
netflix['duration_int'] = pd.to_numeric(netflix['duration'], errors='coerce')
# Rellenar los valores nulos con 0
netflix['duration_int'].fillna(0, inplace=True)
# Convertir la columna a tipo int
netflix['duration_int'] = netflix['duration_int'].astype(int)
```

```python
peliculasyseries = pd.concat([amazon,disney,hulu,netflix],axis=0)
```

```python
peliculasyseries= peliculasyseries.drop(columns='duration')
```

```python
peliculasyseries
```

# Aqui ya estarian listos los dataset de las plataformas con las modificaciones.

```python
# creo variables de cada dataset de la carpeta ratings

ratings1 = pd.read_csv('./MLOpsReviews/ratings/1.csv')
ratings2 = pd.read_csv('./MLOpsReviews/ratings/2.csv')
ratings3 = pd.read_csv('./MLOpsReviews/ratings/3.csv')
ratings4 = pd.read_csv('./MLOpsReviews/ratings/4.csv')
ratings5 = pd.read_csv('./MLOpsReviews/ratings/5.csv')
ratings6 = pd.read_csv('./MLOpsReviews/ratings/6.csv')
ratings7 = pd.read_csv('./MLOpsReviews/ratings/7.csv')
ratings8 = pd.read_csv('./MLOpsReviews/ratings/8.csv')
```

```python
# concateno todos los DF para crear uno solo

ratings = pd.concat([ratings1,ratings2,ratings3,ratings4,ratings5,ratings6,ratings7,ratings8],axis=0)
```

```python
ratings
```

```python
# cambio el fotmato de fecha de timestamp al formato solicitado

ratings['timestamp'] = pd.to_datetime(ratings['timestamp'], unit='s')
```

```python
# me quedo solo con la parte de la fecha del campo date

ratings['date'] = ratings['timestamp'].dt.date
```

```python
#elimino el campo timestamp

ratings =ratings.drop(columns='timestamp')
```

```python
ratings
```

```python
#creo el campo score en donde promedio los ratings de cada movieId

ratings['score_mean'] = ratings.groupby('movieId')['rating'].transform('mean').round(2)
```

```python
scores = ratings.rename(columns={'movieId': 'id','rating':'score'})

```

```python
scores
```

```python
scores_score_mean =  scores.drop_duplicates(subset=['id'])

scores_score_mean = scores_score_mean[['id','score_mean']]
```

```python
df_merge = pd.merge(peliculasyseries, scores_score_mean, on='id', how='left')
df_merge2 = pd.merge(peliculasyseries,scores,on='id',how='left')
```
